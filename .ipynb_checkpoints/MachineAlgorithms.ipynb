```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('yearly_data_above_zero.csv')
X = dataset[['Age']] 
#dataset.drop(['Player First Name', 'Player Last Name','Position','Season','Team','League','Average Time on Ice','Awards', 'Injury Time', 'Injury Description'], axis=1)  # Features
y = dataset['Total Time Out This Year']  # Target variable
```

```python
#from sklearn.preprocessing import StandardScaler
#sc_X = StandardScaler()
#sc_y = StandardScaler()
#X = sc_X.fit_transform(X)
#y = y.values.reshape(-1,1)
# = sc_y.fit_transform(y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.333,random_state=42)
```

```python
#desicion tree
from sklearn import tree
ml_model = tree.DecisionTreeRegressor()
ml_model.fit(X_train,y_train)
```

```python
#random forest
from sklearn.ensemble import RandomForestRegressor
ml_model = RandomForestRegressor(max_depth=2,random_state=00)
ml_model.fit(X_train,y_train)
```

```python
#nearest neighbour
from sklearn.neighbors import KNeighborsRegressor
ml_model = KNeighborsRegressor(n_neighbors=2)
ml_model.fit(X_train,y_train)
```

```python
#support vector machine
from sklearn.svm import SVR
ml_model = SVR(kernel='rbf')
ml_model.fit(X_train,y_train)
```

```python
y_pred = ml_model.predict(X_test)
display(y_pred)
display(y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```

```python
import seaborn as sns

sns.heatmap(cm, 
            annot=True,
            fmt='g')
plt.ylabel('Prediction',fontsize=13)
plt.xlabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```

```python

plt.scatter(y_test, y_pred)
plt.xlabel('Reality')
plt.ylabel('Prediction')
plt.title('Actual vs Predicted time out Width')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(y_test, y_pred, c='crimson')
plt.yscale('log')
plt.xscale('log')

p1 = max(max(y_pred), max(y_pred))
p2 = min(min(y_pred), min(y_pred))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('True Values', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.axis('equal')
plt.show()
```
